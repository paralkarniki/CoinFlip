```python
#3DoorMontyProblem
import random
# defining a function for the number of doors and the switch between the doors
def test(doorswitch,numberofdoors=3):
    selecteddoor = random.randint(1,numberofdoors)
    if doorswitch:
        opendoor = 3 if selecteddoor==2 else 2 
        
        possibledoor = [dnum for dnum in range(1,numberofdoors+1)
                       if dnum not in(selecteddoor,possibledoor)]
        selecteddoor=random.choice(possibledoor)
        
        return selecteddoor==1
#     defining a function that gets the door switch return  
    
def trials(numberoftrials,doorswitch,numberofdoors=3):
    possiblewin =0
    for i in range(numberoftrials):
        if trials(doorswitch,numberofdoors):
            possiblewin+=1
        return possiblewin
    
numberofdoors=3
numberoftrials=10000
possiblewin_with_change=trials(numberoftrials,True,numberofdoors)
possiblewin_without_change=trials(numberoftrials,False,numberofdoors)

print("Monty Hall Problem",numberofdoors)
print("With Switch",possiblewin_with_change/numberoftrials)
print("Without Switch",possiblewin_without_change/numberoftrials)
```
